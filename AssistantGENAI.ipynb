```python
import google.generativeai as genai
from google.colab import userdata
```

```python
api_key=userdata.get("GEMINI_API_KEY")
```

```python
client = genai.configure(api_key=api_key)
```

```python
def ai_assistant(user_prompt):
  """
  Sends a prompt to the Gemini model and gets a response.
  """
  model=genai.GenerativeModel('gemini-3-flash-preview')
  response=model.generate_content(user_prompt)
  return response.text
```

 let’s build a Study Assistant using Python that utilizes the Google Gemini API to explain and simplify learning topics based on a given prompt.

```python
user_question="Explian Generative AI in simple terms"
output=ai_assistant(user_question)
print(output)
```
